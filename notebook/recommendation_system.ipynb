```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv(r"C:\Users\Prachi Singh\OneDrive\Desktop\Projects\imdb_movie_dataset.csv")
movies.head(3)
```

```python
movies = movies[['Rank', 'Title','Genre', 'Description','Director','Actors']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies['Title']=movies['Title'].apply(lambda x:x.replace(" ",""))
movies['Genre']=movies['Genre'].apply(lambda x:x.replace(" ",""))
movies['Description']=movies['Description'].apply(lambda x:x.replace(" ",""))
movies['Director']=movies['Director'].apply(lambda x:x.replace(" ",""))
movies['Actors']=movies['Actors'].apply(lambda x:x.replace(" ",""))


```

```python
movies['Tags'] = movies['Title']+movies['Genre']+movies['Description']+movies['Director']+movies['Actors']
movies.head()
```

```python
new_data = movies[['Rank','Title','Tags']]
new_data.head()    
```

```python
new_data['Tags'] = new_data['Tags'].apply(lambda x:x.lower())
new_data['Tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

```

```python
cv = CountVectorizer(max_features = 7000,stop_words ='english')
```

```python
vectors = cv.fit_transform(new_data['Tags']).toarray()
print(vectors)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
sorted(list(enumerate(similarity[0]))
                       ,reverse = True,
                       key=lambda x: x[1])[1:6]
```

```python
def recommend (movie_name):
  
  movie = movie_name.replace(" ", "")
  matches = new_data[new_data['Title'] == movie]
  
  if matches.empty:
        print("Movie not found in dataset")
        return

  movie_index = matches.index[0]
    
  distances =similarity[movie_index]
  movie_list = sorted(list(enumerate(distances))
                        ,reverse = True,
                       key=lambda x: x[1])[1:5]
  for i in movie_list:
       print(new_data.iloc[i[0]].Title)
  
```

```python
recommend('Guardians of the Galaxy')
```
